# Feature Engineering

### Calculate Network Metrics

```python
users  =  nodes.user_id.to_list()
retweet_dict = {}

for user in users:
    sub_df =  edges[edges.Source == user]
    retweeting_list = sub_df.Target.to_list()
    retweet_dict[user] = retweeting_list

import networkx as nx

G = nx.DiGraph(retweet_dict)
    
in_degree = nx.in_degree_centrality(G)
out_degree = nx.out_degree_centrality(G)
eigenvector = nx.eigenvector_centrality(G)
betweenness = nx.betweenness_centrality(G)
pagerank = nx.pagerank(G)

nodes['in_degree'] = nodes['user_id'].map(in_degree)
nodes['out_degree'] = nodes['user_id'].map(out_degree)
nodes['eigenvector'] = nodes['user_id'].map(eigenvector)
nodes['betweenness'] = nodes['user_id'].map(betweenness)
nodes['pagerank'] = nodes['user_id'].map(pagerank)
```

```python
# Save Backup
nodes.to_csv('backup_nodes_with_network_metrics.csv')
```

### Calculate closeness to Hate

```python
edge_with_info = edges.merge(nodes[['user_id','hate', 'suspended']] , left_on='Target', right_on='user_id', how='left')
edge_with_info.shape[0] == edges.shape[0]

connected_to_hate_dict = {}  
connected_to_normal_dict = {}  
connected_to_suspended_dict = {}
users  =  nodes.user_id.to_list()

for user in users:
    sub_df =  edge_with_info[edge_with_info.Source == user]
    hate_list = sub_df.hate.to_list()
    if 'hateful' in hate_list:
        connected_to_hate_dict[user] = 'connected_to_hate'
    else:
        connected_to_hate_dict[user] = 'not_connected_to_hate'
    
    if 'normal' in hate_list:
        connected_to_normal_dict[user] = 'connected_to_normal'
    else:
        connected_to_normal_dict[user] = 'not_connected_to_normal'
    susp_list = sub_df.suspended.to_list()
    
    if True in susp_list:
        connected_to_suspended_dict[user] = 'connected_to_suspended'
    else:
        connected_to_suspended_dict[user] = 'not_connected_to_suspended'

nodes['connected_to_suspended'] = nodes['user_id'].map(connected_to_suspended_dict)
nodes['connected_to_normal'] = nodes['user_id'].map(connected_to_normal_dict)
nodes['connected_to_hate'] = nodes['user_id'].map(connected_to_hate_dict)
```

### Detect Pronoun

```python
bio_list = nodes.bio.to_list()

pronoun = []
specifies_pronoun = []

for bio in bio_list:
    try:
        if 'he/him' in bio.lower():
            pronoun.append('he/him')
            specifies_pronoun.append(True)
        elif 'she/her' in bio.lower():
            pronoun.append('she/her')
            specifies_pronoun.append(True)
        elif 'they/them' in bio.lower():
            pronoun.append('they/them') 
            specifies_pronoun.append(True)
        else:
            pronoun.append('None') 
            specifies_pronoun.append(False)       
            
    except:
        pronoun.append('None')
        specifies_pronoun.append(False)
        
        
nodes['pronoun'] =  pronoun
nodes['pronoun_bool'] =  specifies_pronoun
```

```python
edge_with_info = edges.merge(nodes[['user_id','pronoun_bool']] , left_on='Target', right_on='user_id', how='left')

connected_pronoun_dict = {}  
connected_pronoun_count_dict = {}
connected_count_dict = {}
users = nodes.user_id.to_list()

for user in users:
    sub_df =  edge_with_info[edge_with_info.Source == user]
    pronoun_list = sub_df.pronoun_bool.to_list()
    connected_pronoun_count_dict[user] = sum(pronoun_list)
    connected_count_dict[user] = len(pronoun_list)
    if sum(pronoun_list) >= 1:
        connected_pronoun_dict[user]= True
    else:
        connected_pronoun_dict[user]= False

nodes['connected_to_pronoun'] = nodes['user_id'].map(connected_pronoun_dict)
nodes['count_connected_to_pronoun'] = nodes['user_id'].map(connected_pronoun_count_dict)
nodes['count_connected'] = nodes['user_id'].map(connected_count_dict)
```

```python
edge_with_info = edges.merge(nodes[['user_id','pronoun']] , left_on='Target', right_on='user_id', how='left')

connected_she_count_dict = {}
connected_he_count_dict = {}
connected_they_count_dict = {}

users = nodes.user_id.to_list()

for user in users:
    sub_df =  edge_with_info[edge_with_info.Source == user]
    pronoun_list = sub_df.pronoun.to_list()
    connected_she_count_dict[user] = pronoun_list.count('she/her')
    connected_he_count_dict[user] = pronoun_list.count('he/him')
    connected_they_count_dict[user] = pronoun_list.count('they/them')

nodes['count_connected_to_she'] = nodes['user_id'].map(connected_she_count_dict)
nodes['count_connected_to_he'] = nodes['user_id'].map(connected_he_count_dict)
nodes['count_connected_to_they'] = nodes['user_id'].map(connected_they_count_dict)
```

```python
nodes['count_connected_to_pronoun_rel'] = nodes['count_connected_to_pronoun']/nodes['count_connected']*100
nodes['count_connected_to_she_rel'] = nodes['count_connected_to_she']/nodes['count_connected']*100
nodes['count_connected_to_he_rel'] = nodes['count_connected_to_he']/nodes['count_connected']*100
nodes['count_connected_to_they_rel'] = nodes['count_connected_to_they']/nodes['count_connected']*100
```
